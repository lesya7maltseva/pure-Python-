**Примечание**

Не все файлы подходящие по названию, будут подходить по содержанию. Там может оказаться лишний столбец, например. Ориентируйтесь на столбцы из чека выше - это то, что вас интересует. Остальные столбцы можно просто отбросить.

**Важно**: разделителем файла на выходе должна быть запятая.

```python
def process_files(src_folder, dest_folder):
  import re
  import glob

  #получаем список всех файлов:
  files = glob.glob(src_folder + '/*.csv')

  #делаем строку из имен файлов:
  str_f = ''
  for i in files:
    str_f += i

  # используя регулярку выбираем файлы, подходящие под условия:
  pattern = src_folder + '/' + '\d{4}-\d{2}-\d{2}-\d{2}-\d{2}-\d+.csv'

  matches_filenames = re.findall(pattern,str_f)
  matches_filenames = sorted(matches_filenames)

  #считываем и редактируем
  prep = []
  table_head = 'date,product,store,cost\n'
  for file in matches_filenames:
    with open (file,'r') as f:
      for line in f:
        if line.endswith('cost\n'): continue
        elif line.startswith('2023'): prep.append(line.replace(';', ','))
        else: prep.append(line[2:].replace(';', ','))

  file_path = dest_folder + '/' + 'combined_data.csv'
  with open (file_path, 'w+') as c:
    c.write(table_head)
    c.writelines(prep)


  ...


src_folder = 'reports-main'
dest_folder = 'comb_reports'
process_files(src_folder, dest_folder)
```
